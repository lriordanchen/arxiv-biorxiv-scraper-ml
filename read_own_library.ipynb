```python
import os
import sys

ArticleData = []

currentwd = os.getcwd()
for file in os.listdir(currentwd+"/swoger_library"):
    if file.endswith(".pdf"):
        pdfname = [file] 
        ArticleData.append(pdfname)
        print(os.path.join("/swoger_library", file))
#flatten data
ArticleData = sum(ArticleData, [])
```

```python
import PyPDF2
from PyPDF2 import PdfFileWriter, PdfFileReader



swoger_lib['body'] = ''
for ix in range(len(swoger_lib.pdf_name)):
    pdf_filename=os.path.join(currentwd+r'\swoger_library',swoger_lib.pdf_name.loc[ix])
    fp = open(pdf_filename,"rb")
    pdfFile = PdfFileReader(fp)
    if not pdfFile.isEncrypted:
        try:
            swoger_lib['body'].loc[ix] = PdfConverter(file_path=pdf_filename).convert_pdf_to_txt().replace('\n',' ')
            swoger_lib.to_pickle('swoger_lib_withbody.pkl')
            print('k')
        except:
            print('missed a file, other error')
            pass
    else:
        print('missed a file, encrypted')
        pass
```
